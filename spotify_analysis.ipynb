# top 5 most popular artists

```python
top_five_artists=spotify_data.groupby('Artists').count().sort_values(by="Title",ascending=False)["Title"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

# Top 5 loudest tracks

```python
top_loudest_tracks=spotify_data[['Loudness','Title']].sort_values(by="Loudness")
top_loudest_tracks.head()
```

# top danceability song

```python
top_danceable_song=spotify_data[['Danceability','Title']].sort_values(by="Danceability",ascending=False)[:5]
# top_danceable_song=top_danceable_song.drop_duplicates()
top_danceable_song
```

# top instrumental tracks

```python
top_instrumental_tracks=spotify_data[['Instrumentalness','Title']].sort_values(by="Instrumentalness",ascending =False)[:5]
top_instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x='Instrumentalness',data=top_instrumental_tracks,autopct='%1.2f%%',labels=top_instrumental_tracks.Title)
plt.title("Top 10 Instrumental Tracks")
plt.show()
```
